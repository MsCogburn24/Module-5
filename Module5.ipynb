```python
#-------------------------------------------
# Program Name: Average Rainfall Calculator
# Author: Stephanie Cogburn
# Date: 18/08/2024
#-------------------------------------------
# Pseudocode:
#   1. Prompt the user to enter the number of years
#   2. Initialize variables to store the total rainfall and the number of months
#   3. Use a nested loop:
#       - Outer loop: Iterate over each year
#       - Inner loop: Iterate over each month (using month names for clarity)
#       - For each month, prompt the user to enter the rainfall in inches and add it to the total rainfall
#   4. After all iterations, calculate the total number of months
#   5. Calculate the average rainfall per month
#   6. Display the total number of months, total inches of rainfall, and average rainfall per month
#-------------------------------------------
# Program Inputs: number_of_years, rainfall_per_month
# Program Outputs: total_months, total_rainfall, average_rainfall
#-------------------------------------------

def average_rainfall_calculator():
    # Input
    number_of_years = int(input("Enter the number of years: "))

    total_rainfall = 0
    total_months = 0

    # List of month names
    months = ["January", "February", "March", "April", "May", "June",
              "July", "August", "September", "October", "November", "December"]

    # Nested loop to collect data
    for year in range(1, number_of_years + 1):
        print(f"\nYear {year}:")
        for month in months:
            rainfall = float(input(f"Enter the rainfall for {month} (in inches): "))
            total_rainfall += rainfall
            total_months += 1

    # Calculate the average rainfall per month
    average_rainfall = total_rainfall / total_months

    # Output
    print("\n--- Rainfall Report ---")
    print(f"Total number of months: {total_months}")
    print(f"Total inches of rainfall: {total_rainfall:.2f}")
    print(f"Average rainfall per month: {average_rainfall:.2f} inches")

# Run the program
average_rainfall_calculator()
```

```python
#-------------------------------------------
# Program Name: Book Club Points Calculator
# Author: Stephanie Cogburn
# Date: 18/08/2024
#-------------------------------------------
# Pseudocode:
#   1. Prompt the user to enter the number of books purchased this month
#   2. Use a series of conditional statements to determine the number of points based on the number of books purchased:
#       - 0 books: 0 points
#       - 2 books: 5 points
#       - 4 books: 15 points
#       - 6 books: 30 points
#       - 8 or more books: 60 points
#   4. Display the number of points awarded or an appropriate message if points are not defined.
#-------------------------------------------
# Program Inputs: number_of_books
# Program Outputs: points_awarded or a message if points are not defined
#-------------------------------------------

def book_club_points_calculator():
    # Input
    number_of_books = int(input("Enter the number of books purchased this month: "))

    # Determine points awarded
    if number_of_books == 0:
        points_awarded = 0
    elif number_of_books == 2:
        points_awarded = 5
    elif number_of_books == 4:
        points_awarded = 15
    elif number_of_books == 6:
        points_awarded = 30
    elif number_of_books >= 8:
        points_awarded = 60
    else:
        points_awarded = "not defined or specified"  # Handle cases where books purchased are not 0, 2, 4, 6, or 8+
        print("Please enter 0, 2, 4, 6, or 8+ books.")

    # Output
    print(f"Points awarded: {points_awarded}")

# Run the program
book_club_points_calculator()
```
